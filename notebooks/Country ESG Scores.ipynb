```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import os
```

```python
%cd ..
%cd data

esg_ind = pd.read_csv('country_esg_indicators.csv')
esg_ind.head()
```

```python
esg_ind.tail(10)
```

```python
esg_ind = esg_ind.iloc[:-5]
```

```python
esg_ind.sort_values(by=['Series Name', 'Country Name'], inplace = True)
esg_ind = esg_ind.reset_index(drop = True)
esg_ind.head()
```

```python
print(esg_ind['2023 [YR2023]'].unique())
```

```python
esg_ind = esg_ind.drop('2023 [YR2023]', axis = 1)
```

```python
esg_ind.rename(columns={'2018 [YR2018]': '2018',
                        '2019 [YR2019]': '2019',
                        '2020 [YR2020]': '2020',
                        '2021 [YR2021]': '2021',
                        '2022 [YR2022]': '2022',
                        }, inplace = True)
esg_ind.head()
```

```python
esg_ind = esg_ind.replace('..', np.nan)
esg_ind.iloc[56:58]
```

```python
i = 0
num_inds = esg_ind['Series Code'].nunique()
num_ctrys = esg_ind['Country Code'].nunique()
years = []
while i < num_inds:
    upper = i * num_ctrys
    lower = (i + 1) * num_ctrys
    subset = esg_ind.iloc[upper:lower]
    search = True
    yr = 2022
    while search:
        if subset[str(yr)].isna().sum() == 0:
            years.append(str(yr))
            search = False
        else:
            yr -= 1

    i += 1
print(years)
```

```python
esg_clean = pd.DataFrame()
esg_clean['Country Name'] = esg_ind['Country Name'].unique()
esg_clean['Country Code'] = esg_ind['Country Code'].unique()
esg_clean.head()
```

```python
i = 0
col_names = esg_ind['Series Name'].unique()
while i < num_inds:
    upper = i * num_ctrys
    lower = (i + 1) * num_ctrys
    subset = esg_ind.iloc[upper:lower]

    name = col_names[i]
    esg_clean[name] = list(subset[years[i]])

    i += 1
esg_clean.head()
```

```python
for name in col_names:
    esg_clean[name] = esg_clean[name].apply(pd.to_numeric)
print(esg_clean.dtypes)
```

```python
school_enrlmt = 'School enrollment, primary and secondary (gross), gender parity index (GPI)'
esg_scaled[school_enrlmt] = abs(1-esg_scaled[school_enrlmt])
esg_scaled.head()
```

```python
scaler = preprocessing.MinMaxScaler()
esg_scaled = scaler.fit_transform(esg_clean.iloc[:,2:])
esg_scaled = pd.DataFrame(esg_scaled, columns = col_names)
esg_scaled.head()
```

```python
neg_fts = ['CO2 emissions (metric tons per capita)',
           'Cause of death, by communicable diseases and maternal, prenatal and nutrition conditions (% of total)',
           'Level of water stress: freshwater withdrawal as a proportion of available freshwater resources',
           'Methane emissions (metric tons of CO2 equivalent per capita)',
           'Mortality rate, under-5 (per 1,000 live births)',
           'Nitrous oxide emissions (metric tons of CO2 equivalent per capita)',
           'PM2.5 air pollution, mean annual exposure (micrograms per cubic meter)',
           'School enrollment, primary and secondary (gross), gender parity index (GPI)',
           'Tree Cover Loss (hectares)'
           ]

for ft in neg_fts:
    esg_scaled[ft] = -1*esg_scaled[ft]



esg_scaled.head()
```

```python
esg_cats = {'E': ['CO2 emissions (metric tons per capita)',
                  'Level of water stress: freshwater withdrawal as a proportion of available freshwater resources',
                  'Methane emissions (metric tons of CO2 equivalent per capita)',
                  'Nitrous oxide emissions (metric tons of CO2 equivalent per capita)',
                  'PM2.5 air pollution, mean annual exposure (micrograms per cubic meter)',
                  'Renewable energy consumption (% of total final energy consumption)',
                  'Terrestrial and marine protected areas (% of total territorial area)',
                  'Tree Cover Loss (hectares)'
                  ],
            'S': ['Access to clean fuels and technologies for cooking (% of population)',
                  'Cause of death, by communicable diseases and maternal, prenatal and nutrition conditions (% of total)',
                  'Government expenditure on education, total (% of government expenditure)',
                  'Life expectancy at birth, total (years)',
                  'Mortality rate, under-5 (per 1,000 live births)'
                  ],
            'G': ['Control of Corruption: Estimate',
                  'Economic and Social Rights Performance Score',
                  'Government Effectiveness: Estimate',
                  'Political Stability and Absence of Violence/Terrorism: Estimate',
                  'Proportion of seats held by women in national parliaments (%)',
                  'Regulatory Quality: Estimate',
                  'Rule of Law: Estimate',
                  'School enrollment, primary and secondary (gross), gender parity index (GPI)',
                  'Strength of legal rights index (0=weak to 12=strong)',
                  'Voice and Accountability: Estimate'
                  ]}
```

```python
esg_scores = []
for i in range(num_ctrys):
    components = []
    for cat in esg_cats:
        sub_score = 0
        for stat in esg_cats[cat]:
            sub_score += (esg_scaled[stat])[i]
        components.append(sub_score)
    avg = sum(components)/len(components)
    esg_scores.append(avg)
```

```python
esg_clean['ESG Score'] = esg_scores
esg_clean = esg_clean.sort_values(by = 'ESG Score', ascending=False)
esg_final = esg_clean[['Country Name', 'ESG Score']]
esg_final
```
